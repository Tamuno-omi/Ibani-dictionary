```python
import pandas as pd
import numpy as np
import re
from sqlalchemy import create_engine
import json
```

```python
# read in data
path ='Merged file.xlsx'
ib = pd.read_excel(path,index_col=None)
```

```python
ib.head()
```

```python
ib.tail()
```

```python
# rename columns
ib.rename(columns = {'Ibanị':'Ibani','PoS':'Pos'}, inplace = True)
```

```python
ib.info()
```

```python
# check for null 
ib.isnull().sum()
```

```python
# drop null rows in Ibani column
ib.dropna(subset=['Ibani'],inplace = True)
```

```python
# remove rows with numeric value in Ibani
ib = ib[ib['Ibani'].apply(lambda x: str(x).isalpha())]
```

```python
ib.head()
```

```python
# columns in pos with number
ib[ib['Pos'].apply(lambda x: str(x).isnumeric())]
```

```python
# check for null in Pos column
ib[ib['Pos'].isnull()]
```

```python
# check for null in Gloss column
ib[ib['Pos'].isnull()]
```

```python
# replace null rows in Pos column
ib['Pos'] = ib['Pos'].fillna('nill')
# replace null rows in Gloss column
ib['Gloss'] = ib['Gloss'].fillna('nill')
```

```python
# replace pos rows with numbers with values in Gloss 
ib.Pos[ib['Pos'].apply(lambda x: str(x).isnumeric())] = ib.Gloss
```

```python
ib.head()
```

```python
# fill null values in Ibani dataframe 
ib = ib.fillna('')
```

```python
# combine unnamed columns
ib['Combined'] = ib.loc[:,ib.columns.str.startswith('Unnamed:')].apply(lambda x:' '.join(x.values.astype(str)), axis = 1)
```

```python
ib.head()
```

```python
# replace part of speech values found in gloss column
ib['Gloss'] = ib['Gloss'].replace(to_replace=r'^\w*\W$', value='', regex=True)
```

```python
# combine gloss and combined column
ib["Glossary"] = ib["Gloss"] +" "+ ib["Combined"]
```

```python
ib_clean = ib[['Ibani','Pos','Glossary']]
ib_clean
```

```python
ib['Pos'].unique()
```

```python
# check for invalid values in pos
ib_clean['Pos'][ib_clean['Pos'].str.match(r"^\w*\w$")==True]
```

```python
# list of invalid values in Pos column
ib_clean['Pos'].loc[ib_clean['Pos'].str.match(r"^\w*\w$")==True].values.astype(str)
```

```python
# view pos row with invalid value
ib_clean[ib_clean['Pos'] == '9l9']
```

```python
# clean row with wrong pos
#ib_clean.Pos[ib_clean['Pos'] == '9l9'] = 'v.p'
ib_clean.loc[ib_clean.Pos == '9l9', 'Pos'] = 'v.p'
```

```python
ib_clean.loc[ib_clean['Pos'].str.match(r"^\w*\w$")==True]
```

```python
# get the ibani words truncated into the pos column and concatenate to the Ibani column
mask = ib_clean['Pos'].str.match(r"^\w*\w$")
ib_clean["Ibani"][mask] = ib_clean["Ibani"]+' '+ ib_clean["Pos"]
```

```python
ib_clean[ib_clean['Pos'] == 'numó']
```

```python
# replace wrong pos values with nill in pos column
X = ['pron.', 'excl.', 'n.', 'a.','v.','dem.', 'adv.', 'n.pl.','num.',
     'v.cs.','intro.', 'v.t.','v.reci','v.p.', 'p.n.','v.i.','v./n.',
     'v./n.', 'v.a.', 'a./ad', 'v.aux', 'n.p.', 'part.', 'neg.', 'conj.','pp.','1.conj.', '2.intro.']
ib_clean.loc[~ib_clean.Pos.isin(X), 'Pos'] = 'nill'
```

```python
# rename columns
ib_clean.rename(columns = {'Glossary':'Meaning'}, inplace = True)
ib_clean.head()
```

```python
def save_data(df):
    '''saves dataframe to sql databases
    input
    
    df: cleaned dataframe    
    '''
    engine = create_engine('sqlite:///IbaniDictionary.db')
    df.to_sql('Ibani', engine, index=False, if_exists='replace')
```

```python
#save dataframe to sql
save_data(ib_clean)
#save dataframe to json
Ibani = ib_clean.to_json('IbaniDictionary.json',orient="index")
```
